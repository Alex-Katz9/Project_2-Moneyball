```python
df_filtered = analysis_df[analysis_df['PA'] >= 50]
```

```python
df_filtered_2 = df_filtered.apply(pd.to_numeric, errors='coerce')
df_filtered = df_filtered.dropna()
df_filtered
```

```python
test_df = df_filtered
test_df['Orig_Pos'] = test_df.iloc[:, 19]


test_df.drop(test_df.columns[19], axis=1, inplace=True)

test_df
```

```python
test2_df = test_df
```

```python
for i, row in test2_df.iterrows():
    string = str(row['Orig_Pos'])
    res = [int(j) for j in string if j.isdigit()]
    if res:
        test2_df.loc[i,'Orig_Pos'] =res[0]   
    else:
        test2_df.drop([i])
```

```python
test2_df
```

```python
test3_df = test2_df[test2_df['Orig_Pos'] >= 2]
test3_df
```

```python
type_df = test3_df.copy()
type_df['Position'] = test3_df['Orig_Pos']
type_df
```

```python
type_df['Position'].nunique()
```

```python
positions_df = pd.get_dummies(type_df['Position'])
positions_df
```

```python
EA_df = pd.concat([type_df,positions_df], axis = 1)
EA_df
```

```python
EA_df.drop(['Orig_Pos'], axis=1, inplace=True)
EA_df.drop(['Position'], axis=1, inplace=True)
EA_df.drop(['BB'], axis=1, inplace=True)
# EA_df.drop(['R'], axis=1, inplace=True)
EA_df.drop(['CS'], axis=1, inplace=True)
EA_df.drop(['TB'], axis=1, inplace=True)
EA_df.drop(['HBP'], axis=1, inplace=True)
EA_df.drop(['SH'], axis=1, inplace=True)
EA_df.drop(['SF'], axis=1, inplace=True)
EA_df.drop(['PA'], axis=1, inplace=True)
# EA_df.drop(['BA'], axis=1, inplace=True)
# EA_df.drop(['OPS'], axis=1, inplace=True)

EA_df
```
